```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import SimpleITK as sitk
from myshow import myshow, myshow3d
```

```python
rgb_img = sitk.ReadImage("t009_resized_small.tif")
myshow(rgb_img, "Original Image")
```

```python
img = sitk.VectorIndexSelectionCast(rgb_img,1)
myshow(img, "Green Image")
```

```python
feature_img = sitk.GradientMagnitudeRecursiveGaussian(img, sigma=0.00001)
myshow(feature_img)
```

```python
ws_img = sitk.MorphologicalWatershed(feature_img, level=10000, markWatershedLine=True, fullyConnected=True)
myshow(sitk.LabelToRGB(ws_img), "Watershed Over Segmentation")
```

```python
seg = sitk.ConnectedComponent(ws_img!=ws_img[340, 340])
myshow(seg, "Connected Component")
myshow(sitk.LabelOverlay(img, seg), "Foreground Components")
```

```python
filled = sitk.BinaryFillhole(seg!=0)
d = sitk.SignedMaurerDistanceMap(filled, insideIsPositive=False, squaredDistance=False, useImageSpacing=False)
myshow(d, "Inside Distance Map")
```
